```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#Prepare the dataset
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
print(df.head())
```

```python
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
```

```python
df.shape
```

```python
print(df.isnull().sum())
df = df.dropna()
```

```python
X = df.drop("Purchased", axis=1)  # Independent variables
Y = df['Purchased']                 # Dependent variable
```

```python
print("Shape of my X= ",X.shape)
print("Shape of my X= ",Y.shape)
```

```python
#Train Test Split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=20)
```

```python
X_train.shape
X_test.shape
```

```python
X_train.shape
```

```python
print("Shape of my X_train= ",X_train.shape)
print("Shape of my X_test= ",X_test.shape)
print("Shape of my Y_train= ",Y_train.shape)
print("Shape of my Y_test= ",Y_test.shape)
```

```python
SC=StandardScaler()
SC.fit(X_train)
X_train=SC.transform(X_train)
X_test=SC.transform(X_test)
```

```python
X_train
```

```python
lr=LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
lr.coef_
```

```python
y_pred = lr.predict(X_test)
print("Accuracy:", accuracy_score(Y_test, y_pred))
```

```python
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred))
```

```python
print("Classification Report:\n", classification_report(Y_test, y_pred))
```
